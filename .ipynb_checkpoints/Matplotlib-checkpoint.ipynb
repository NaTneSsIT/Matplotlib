## Pylot API

```python
#demo
x=np.linspace(0,10,100)
plt.plot(x,x**2,linestyle="dashed",label="x$^2$");
plt.plot(x,np.sin(x),label="Sin(x)");
plt.xlabel("X");
plt.ylabel("Y");
plt.title("Chart x$^2$ and Sin(x)");
plt.axis("tight")
plt.xlim([0,8])
#plt.axis([xmin,xmax,ymin,ymax])
plt.legend();
```

## Object-Oriented API

```python
#demo
x=[1,2,3,4]
y=[4,8,6,7]
fig,ax=plt.subplots(figsize=(8,7))
ax.plot(x,y,label="xy");
ax.set(title="Demo",xlabel="Variable x",ylabel="Variable y");
plt.legend();
```

### Line

```python
#Demo
fig,ax=plt.subplots()
ax.plot(x, color="red");
ax.set(title="Line");
```

### Scatter

##### O1:Pyplot API 

```python
#demo
plt.scatter(x,x**2,cmap="viridis",s=1000,alpha=0.3)
plt.xlim(0,1)
```

##### O2:O-O API

```python
r=np.random.RandomState(0)
```

```python
x=r.rand(100)
y=r.rand(100)
fig,ax=plt.subplots()
sizes=1000*r.rand(100)
colors=r.rand(100)
demo=ax.scatter(x,y,s=sizes,c=colors,cmap="magma",alpha=0.4);
ax.set(title="Balloons");
fig.colorbar(demo);
```

### Bar

```python
#prepare data (dictionary)
quantity_class={"12A13":10,
                "12A14":20,
                "12A15":15}
   
fig,ax=plt.subplots();
ax.bar(quantity_class.keys(),quantity_class.values(),color="wheat",label="Quantity")
ax.set(xlabel="Name",ylabel="Quantity",title="quantity_class");
plt.legend();

   
```

```python
#hozirontal chart
meat_prices={"Chicken":10,
                "Pork":20,
                "Beff":15}
fig,ax=plt.subplots();
ax.set(xlabel="Quantity",ylabel="Name",title="quantity_class");
ax.barh(list(meat_prices.keys()),list(meat_prices.values()),color="wheat",label="Quantity")
plt.legend();
```

### Histogram

```python

dog_height=np.random.normal(60,10,150)
plt.hist(dog_height,color="wheat",label="height")
plt.legend();
```

### Subplots()

```python
#option 1
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x)
ax2.scatter(r.rand(100),r.rand(100),c=r.rand(100),cmap="magma",alpha=0.4);
ax3.bar(quantity_class.keys(),quantity_class.values(),color="wheat",label="Quantity")
ax4.hist(dog_height,color="wheat",label="height")
```
